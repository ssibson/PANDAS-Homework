```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
Path.cwd()
```

```python
# Setting file paths
whale_data = Path("../Starter-Files/whale_returns.csv")
algo_data = Path("../Starter-Files/algo_returns.csv")
sp500_data = Path("../Starter-Files/sp500_history.csv")

# Reading the CSVs and setting the `date` column as a datetime index to the DataFrame
whale_df = pd.read_csv(whale_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_df = pd.read_csv(algo_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_df = pd.read_csv(sp500_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
```

```python
#Reading Whale Data

whale_df
```

```python
# Checking for NaNs within whales data

whale_df.isnull().sum()
```

```python
# Cleaning nulls from whales by dropping

whale_df = whale_df.dropna().copy()
whale_df
```

```python
# Checking removal of NaNs from whales

whale_df.isnull().sum()
```

```python
#Reading algo data
algo_df
```

```python
# Checking for NaNs within algo data

algo_df.isnull().sum()
```

```python
# Cleaning nulls from algo_df by dropping

algo_df = algo_df.dropna().copy()
algo_df
```

```python
# Checking removal of NaNs from algo data

algo_df.isnull().sum()
```

```python
#Reading S&P 500 data
sp500_df
```

```python
# Retrieving S&P 500 data types

sp500_df.dtypes
```

```python
# Cleaning '$' From Close Price
sp500_df['Close'] = sp500_df['Close'].str.replace('$','')
sp500_df['Close']
```

```python
# Checking Close data type
sp500_df['Close'].dtype
```

```python
# Casting Close Series as a Float
sp500_df['Close'] = sp500_df['Close'].astype('float')
sp500_df['Close'].dtype
```

```python
# Confirming Data type Fix

sp500_df.dtypes
```

```python
# Sorting dataframe in ascending
sp500_df.sort_index(inplace=True)
```

```python
# Calculate the daily return using the 'shift()' function
daily_returns = (sp500_df - sp500_df.shift(1)) / sp500_df.shift(1)
daily_returns
```

```python
# Cleaning nulls from daily returns by dropping

daily_returns = daily_returns.dropna().copy()
daily_returns
```

```python
#Rename 'Close' Column to 'S&P 500'
daily_returns.columns = ['S&P 500']
daily_returns.head()
```

```python
sp500_df
```

```python
# Concatenating all DataFrames to one
combined_df = pd.concat([whale_df, algo_df, daily_returns], axis="columns", join="inner")
combined_df.head()
```

```python
##### PORTFOLIO ANALYSIS #####
```

```python
# Plot daily returns

combined_df.plot(figsize=(20, 10), title='Daily Returns')
```

```python
# Using the `cumprod` function to calculate cumulative returns

cumulative_returns = (1 + combined_df).cumprod()
cumulative_returns.plot(figsize=(20, 10), title='Cumulative Returns')
```

```python
# Visualizing data risk via Box Plot

combined_df.plot.box(figsize=(20, 10), title='Portfolio Risk')
```

```python
# Daily Standard Deviations

daily_std = combined_df.std()
daily_std.head(7)
```

```python
# Identifying the portfolio with the most risk

daily_std = daily_std.sort_values(ascending=False)
daily_std.head(7)
```

```python
# Determining which portfolios are riskier than the S&P 500
sp500risk = daily_std.loc['S&P 500']
daily_std > sp500risk
```

```python
# Calculating the annualized standard deviation
annualized_std = daily_std * np.sqrt(252)
annualized_std.head(7)
```

```python
##### ROLLING STATISTICS #####
```

```python
# Calc and plot 21-Day Rolling Standard Deviation

combined_df.rolling(window=21).std().plot(figsize = (20, 10), title = '21-Day Rolling Standard Deviation')
```

```python
# Constructing a correlation table (with Pearson Method)

correlation = combined_df.corr(method="pearson")
correlation
```

```python
BHI_covariance=combined_df['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(combined_df["S&P 500"])
variance=combined_df["S&P 500"].rolling(60).var()
(BHI_covariance/variance).plot(figsize=(20,10),title="Berkshire Hathaway Inc. Beta")

#BHI_covariance = combined_df['BERKSHIRE HATHAWAY INC'].cov(combined_df['S&P 500'])
#variance = combined_df['S&P 500'].var()
#BHI_beta = BHI_covariance / variancefd
#BHI_rolling = combined_df['BERKSHIRE HATHAWAY INC'].rolling(21).cov(combined_df['S&P 500'])
#BHI_rolling.plot(figsize=(20, 10), title='Berkshire Hathaway Inc. Beta')
```

```python
# Calculating and plotting a rolling window using the exponentially weighted moving average. 

combined_df.ewm(halflife=21).std().plot(figsize=(20, 10), title = 'Exponentially Weighted Average')
```

```python
# Calculating annualized Sharpe Ratios

sharpe_ratios = (combined_df.mean() * 252) / (daily_std * np.sqrt(252))
sharpe_ratios
```

```python
# Visualizing the sharpe ratios as a bar plot

sharpe_ratios.plot.bar(figsize = (20, 10), title = 'Sharpe Ratios')
```

```python
# READING IN FFIV Stock
FFIV_data = Path("../Starter-Files/FFIV_Data.csv")
ffiv_df = pd.read_csv(FFIV_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
ffiv_df.head()
```

```python
# READING IN COST Stock
COST_data = Path("../Starter-Files/COST_Data.csv")
cost_df = pd.read_csv(COST_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
cost_df.head()
```

```python
# READING IN PANW Stock
PANW_data = Path("../Starter-Files/PANW_data.csv")
panw_df = pd.read_csv(PANW_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
panw_df.head()
```

```python
# READING IN PYPL Stock
PYPL_data = Path("../Starter-Files/PYPL_data.csv")
pypl_df = pd.read_csv(PYPL_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
pypl_df.head()
```

```python
# Concatenate all stocks into a single DataFrame

combined_portfolio_df = pd.concat([cost_df, ffiv_df, panw_df, pypl_df], axis="columns", join="inner")
combined_portfolio_df.sort_index(inplace=True)
combined_portfolio_df.columns = ["COST", "FFIV", "PANW", "PYPL"]
combined_portfolio_df.head()
```

```python
# Sorting dataframe in ascending
combined_portfolio_df.sort_index(inplace=True)
```

```python
# Calculate the daily return using the 'shift()' function
daily_pf_returns = (combined_portfolio_df - combined_portfolio_df.shift(1)) / combined_portfolio_df.shift(1)
daily_pf_returns
```

```python
# Cleaning nulls from DataFrame by dropping
daily_pf_returns = daily_pf_returns.dropna().copy()
daily_pf_returns.head()
```

```python
# Calculate weighted portfolio returns
weights = [1/4, 1/4, 1/4, 1/4]
portfolio_returns = daily_pf_returns.dot(weights)
portfolio_returns
```

```python
# Adding Custom Portfolio to earlier Combined DataFrame

final_df = pd.concat([whale_df, algo_df, daily_returns],axis="columns", join="inner")
final_df["Custom"] = portfolio_returns
final_df.sort_index(inplace=True)
final_df
```

```python
# Cleaning nulls from combined DataFrame by dropping
final_df = final_df.dropna().copy()
final_df.head()
```

```python
# Assessing risk
final_volatility = final_df.std() * np.sqrt(252)
final_volatility.sort_values(inplace=True)
final_volatility
```

```python
# Calculating and plotting a rolling window using the exponentially weighted moving average. 

final_df.ewm(halflife=21).std().plot(figsize=(20, 10), title = '21-Day Rolling Standard Deviation')
```

```python
# Calculating annualized Sharpe Ratios

final_sharpe_ratios = (final_df.mean() * 252) / (final_df.std() * np.sqrt(252))
final_sharpe_ratios
```

```python
# Visualizing the sharpe ratios as a bar plot

final_sharpe_ratios.plot.bar(figsize = (20, 10), title = 'Sharpe Ratios including Custom Portfolio')
```

```python
# Constructing a final correlation table (with Pearson Method)

final_correlation = final_df.corr(method="pearson")
final_correlation
```

```python
# Calculating and plotting Custom portfolio beta

final_covariance=final_df['Custom'].rolling(window=60).cov(final_df["S&P 500"])
final_variance=combined_df["S&P 500"].rolling(60).var()
(final_covariance/final_variance).plot(figsize=(20,10),title="Custom Portfolio Beta")
```
